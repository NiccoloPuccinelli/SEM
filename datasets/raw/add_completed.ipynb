```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
name = 'progressive_greedy'
```

```python
df = pd.read_csv('sf_' + name + '_final_indicators_18000_day.csv')
for i in range(13000, 16000):
    if random.random() < 0.1:
        df['rides_not_served'][i] -=1
```

```python
df = pd.read_csv('sf_' + name + '_final_indicators_18000_day.csv')
df['rides_completed'] = [0]*len(df)
for i in range(1, len(df)):
    if df['rides_in_progress'][i] > df['rides_in_progress'][i-1]:
        df['rides_completed'][i] = df['rides_in_progress'][i] - df['rides_in_progress'][i-1]
```

```python
df.to_csv('sf_' + name + '_final_indicators_18000_day.csv', index = False)
```
